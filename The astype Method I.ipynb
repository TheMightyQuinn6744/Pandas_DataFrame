```python
#import pandas under the alias pd
import pandas as pd
```

```python
#import data from nba csv into nba
nba = pd.read_csv("nba.csv").dropna(how = "all")
#Overwrite the original missing salary values in the NBA DataFrame
nba["Salary"] = nba["Salary"].fillna(0)
nba
```

```python
#If you have a missing value in a numeric column, pandas import that numeric column as floating points
#We can now convert this age column to store integer values using the "as type" method
nba["Age"].hasnans
nba["Age"].astype("int")
#this may not work with specific data types
nba["Age"].astype(int)
```

```python
#Overwrite the original Age series and then check the DataFrame
nba["Age"] = nba["Age"].astype(int)
nba["Age"] = nba["Age"].astype(int)
```

```python
nba["Salary"].hasnans
nba
```
